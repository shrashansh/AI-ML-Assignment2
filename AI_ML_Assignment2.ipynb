```python
import googletrans
r = sr.Recognizer()
inp = sr.AudioFile('/content/speechmodi.wav')

with inp as file:
  audio = r.record(file)

testtext = r.recognize_google(audio, language="hi")
print(testtext)
translator = googletrans.Translator()
text = translator.translate(testtext, dest = "en").text
print(text)
```

```python
CUSTOM_VOICE_NAME = "narendramodi"

custom_voice_folder = f"tortoise/voices/{CUSTOM_VOICE_NAME}"
os.makedirs(custom_voice_folder)
for i, file_data in enumerate(files.upload().values()):
  with open(os.path.join(custom_voice_folder, f'{i}.wav'), 'wb') as f:
    f.write(file_data)
```

```python
preset = "high_quality"
voice_samples, conditioning_latents = load_voice("narendramodi")
gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents,
                          preset=preset)
torchaudio.save(f'generated-narendramodi.wav', gen.squeeze(0).cpu(), 24000)
IPython.display.Audio(f'generated-narendramodi.wav')
```
